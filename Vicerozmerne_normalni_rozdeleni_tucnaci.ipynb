Vícerozměrné normální rozdělení - příklad s tučňáky

```python
import numpy as np
import pandas as pd

# Mean vector for [weight, height, waist circumference]
mean_vector = [20, 30, 60]

# Standard deviations for [weight, height, waist circumference]
std_devs = [3, 5, 8]

# Correlation matrix
corr_matrix = [
    [1.0, 0.7, 0.8],
    [0.7, 1.0, 0.6],
    [0.8, 0.6, 1.0]
]

# Convert correlation matrix to covariance matrix
cov_matrix = np.diag(std_devs) @ corr_matrix @ np.diag(std_devs)

# Generate 1000 samples from multivariate normal distribution
data = np.random.multivariate_normal(mean_vector, cov_matrix, size=1000)

# Create a DataFrame for easier analysis
df = pd.DataFrame(data, columns=['Weight (kg)', 'Height (inches)', 'Waist (cm)'])
df
```

```python
# Calculate mean vector, standard deviation vector, and correlation matrix
mean_vec = df.mean()
std_dev_vec = df.std()
correlation_matrix = df.corr()

print("Mean Vector:\n", mean_vec)
print("\nStandard Deviation Vector:\n", std_dev_vec)
print("\nCorrelation Matrix:\n", correlation_matrix)
```
